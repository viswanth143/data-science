```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y= [1,4,9,16]
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Basic Line Plot')
plt.show()
```

```python
x = [1,2,3,4]
y= [1,4,9,16]
plt.plot(x,y, color = 'green', linestyle='solid', marker='o', linewidth = 2,markersize=9) #line style colour and places where to mark

```

```python
#multiple plots
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]
plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y1,color='green')
plt.title('plot 1')
plt.subplot(1,2,2)
plt.plot(x,y1,color='green')
plt.title('plot 2')
```

```python
#barplot
categories = ['A', 'B', 'C', 'D', 'E']
values = [5,7,8,3,6]

#create a bar plot
plt.bar(categories,values,color='orange')
plt.
```

```python
#Histogram which uses to define the distribution of the dataset they divide the data into bins and count number of data points in each bin
data = [1,2,3,4,4,5,6,7,7,7]
plt.hist(data,bins=5, color='green',edgecolor='black')
```

```python
#create a scatter plot
x = [1,2,3,4,5]
y = [1,2,3,6,9]
plt.scatter(x,y, color='blue',marker = 'x')
```

```python
#pie chart
labels = ['A','B','C','D']
SIZE = [30,40,50,60]
colors = ['green','gold','yellow','orange']
explode=(0.2,0,0,0)

plt.pie(SIZE,labels=labels,colors=colors,autopct='%1.1f%%',explode = explode)
```
