```python
#series
import pandas as pd
data =[1,2,3,4,5]
series = pd.Series(data)
print('series \n',series)
```

```python
#CREATE A SERIES FROM DICTIONARY
data = {'a':1, 'b':2, 'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
#create a dataframe from a dictionary of list
data ={
    'Name': ['Ram', 'Krishna', 'Shiva'],
    'Age': [21,22,23],
    'City': ['Ayodhya', 'Dwaraka', 'kasi']
}
df=pd.DataFrame(data)
print(df)
```

```python
#create a dataframe from list of dictionaries.
data =[
    {'Name': 'Ram', 'Age': 32,'City': 'Ayodhya'},
    {'Name': 'krishna', 'Age': 42,'City': 'dwaraka'},
    {'Name': 'Ramkrishna', 'Age': 22,'City': 'Ayodhya'}
]
df=pd.DataFrame(data)
print(df)
```
