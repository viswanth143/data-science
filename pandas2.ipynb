```python
import pandas as pd

# Replace with the correct file path to your CSV
file_path = '/Users/viswanth/downloads/data.csv'

# Load the CSV file into a Pandas DataFrame
df = pd.read_csv(file_path)

df['Sales_fillNA']=df['Sales'].fillna(df['Sales'].mean())
```

```python
#data aggregating and grouping
df.head()
```

```python
grouped_mean = df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
#aggreagting multiple functions
grouped_agg=df.groupby('Region')['Value'].agg(['mean', 'sum', 'count'])
grouped_agg
                                              
```

```python
#merging and joining data frames
df1= pd.DataFrame({'Key':['A', 'B', 'C'], 'Value1': [1, 2, 3]})
df2= pd.DataFrame({'Key':['A', 'B', 'C'], 'Value1': [4, 5, 6]})
pd.merge(df1,df2,on='Key',how='inner')
```
